```python
rows = 3; 
cols = 3; 
```

```python
mat = [[1, 2, 3], 
       [4, 5, 6], 
       [7, 8, 9]]; 
```

```python
def meanColumnWise(mat): 
    print("[ ", end = ""); 
  
    for i in range(rows): 
          
        mean = 0.00; 
  
        sum = 0; 
  
        for j in range(cols): 
            sum = sum + mat[j][i]; 
  
        mean = int(sum /rows); 
        print(mean, end = " "); 
  
    print("]"); 
```

```python
meanColumnWise(mat); 
```

```python
def meanRowWise(mat): 
    print("[ ", end = ""); 
  
    for i in range(cols): 
          
        mean = 0.00; 
  
        sum = 0; 
  
        for j in range(rows): 
            sum = sum + mat[i][j]; 
  
        mean = int(sum /cols); 
        print(mean, end = " "); 
  
    print("]"); 
```

```python
meanRowWise(mat); 
```
